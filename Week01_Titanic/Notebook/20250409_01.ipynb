```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_titanic.csv')
```

```python
df.head()
```

```python
df['Class'].value_counts().sort_index().plot(kind = 'bar')

plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.xticks(rotation = 0)
plt.show()
```

```python
sns.countplot(x = 'Class', data = df)

plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.hist(df['Fare'],bins = 10, edgecolor = 'black')

plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
sns.histplot(df['Fare'], bins = 15, kde = True)

plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()
```

```python
sns.boxplot(x = 'Survived', y = 'Age', hue = 'Gender', data = df)

plt.title('Age Distribution by Survival')
plt.xlabel('Survival Status (0 = Died, 1 = Survived)')
plt.ylabel('Age')
plt.show()
```
