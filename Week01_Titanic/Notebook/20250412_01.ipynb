```python
from my_utils import clean_titanic_data, plot_count, age_category

df = clean_titanic_data('original_titanic.csv')
df.head(10)
```

```python
subset = df[(df['Class'] == 3) & (df['Fare'] > 30) & (df['Survived'] == 1)]
print(subset)
```

```python
subset = df[(df['Age'] < 18) & (df['Fare'] > 20) & (df['Survived'] ==0)]
print(subset)
```

```python
Fare_by_Class = df.groupby('Class')['Fare'].mean()
print(Fare_by_Class)
Age_by_Gender = df.groupby('Gender')['Age'].mean()
print(Age_by_Gender)
Fare_by_Gender = df.groupby('Gender')['Fare'].mean()
print(Fare_by_Gender)
```

```python
Age_and_Fare_by_Gender = df.groupby('Gender').agg({'Age': 'mean','Fare': 'mean'})
print(Age_and_Fare_by_Gender)
```

```python
Fare_by_Class_and_Gender = df.groupby(['Class','Gender'])['Fare'].mean()
print(Fare_by_Class_and_Gender)
```

```python
Fare_by_Class_and_Gender = df.groupby(['Class','Gender'])['Fare'].mean().reset_index()
print(Fare_by_Class_and_Gender)
```

```python
df['AgeGroup'] = df['Age'].apply(age_category)
df.head(10)
```

```python
import numpy as np

conditions = [df['Fare'] < 10, (10 <= df['Fare']) & (df['Fare'] < 30), 30 <= df['Fare']]
choices = ['Low', 'Medium', 'High']

df['FareTier'] = np.select(conditions, choices)
df.head(10)
```
