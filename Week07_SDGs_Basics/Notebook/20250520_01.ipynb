```python
import pandas as pd
```

```python
data = pd.read_csv('global_sustainable_energy.csv')
```

```python
data.info()
```

```python
len(data['Entity'].unique())
```

```python
missing_by_country = data.groupby("Entity").apply(lambda df: df.isnull().sum(), include_groups = False)

missing_by_country["Total Missing"] = missing_by_country.sum(axis = 1)

missing_sorted = missing_by_country.sort_values("Total Missing", ascending = True)

missing_sorted.head(50)
```
