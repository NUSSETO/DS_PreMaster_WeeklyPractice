```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("global_sustainable_energy.csv")
```

```python
countires = ["Brazil", "India", "Mexico", "Thailand"]
data_focus = data[data["Entity"].isin(countires)]
```

```python
plt.figure(figsize = (10, 6))

sns.lineplot(data = data_focus,
             x = "Year",
             y = "Primary energy consumption per capita (kWh/person)",
             hue = "Entity",
             marker = "o")

plt.title("Primary Energy Consumption per Capita (20002020)")

plt.xticks(data_focus["Year"].unique(), rotation = 45)

plt.ylabel("Energy Consumption (kWh/person)")
plt.xlabel("")

plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10, 6))

sns.lineplot(data = data_focus,
             x = "Year",
             y = "gdp_per_capita",
             hue = "Entity",
             marker = "o")

plt.title("GDP per Capita (20002020)")

plt.xticks(data_focus["Year"].unique(), rotation = 45)

plt.ylabel("GDP per Capita (USD)")
plt.xlabel("")

plt.grid()
plt.tight_layout()
plt.show()
```
