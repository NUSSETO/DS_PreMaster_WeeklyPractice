```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("global_sustainable_energy.csv")
```

```python
countires = ["Brazil", "India", "Mexico", "Thailand"]
data_focus = data[data["Entity"].isin(countires)]
```

```python
plt.figure(figsize = (10, 6))

sns.lineplot(data = data_focus,
             x = "Year",
             y = "Access to electricity (% of population)",
             hue = "Entity",
             marker = "o")

plt.title("Electricity Access Over Time (20002020)")

plt.xticks(data_focus["Year"].unique(), rotation = 45)

plt.ylabel("Electricity Access (%)")
plt.xlabel("")

plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10, 6))

sns.lineplot(data = data_focus,
             x = "Year",
             y = "Renewable energy share in the total final energy consumption (%)",
             hue = "Entity",
             marker = "o")

plt.title("Renewable Energy Share in Final Energy Consumption (20002020)")

plt.xticks(data_focus["Year"].unique(), rotation = 45)

plt.ylabel("Renewable Share (%)")
plt.xlabel("")

plt.grid()
plt.tight_layout()
plt.show()
```
