```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("global_sustainable_energy.csv")
```

```python
data_focus = data[data["Entity"] == 'India']
```

```python
plt.figure(figsize = (10, 6))

sns.lineplot(data = data_focus,
             x = "Year",
             y = "Access to electricity (% of population)",
             hue = "Entity",
             marker = "o")

plt.title("Electricity Access Over Time (20002020)")

plt.xticks(data_focus["Year"].unique(), rotation = 45)

plt.ylabel("Electricity Access (%)")
plt.xlabel("")

plt.grid()
plt.tight_layout()
plt.savefig("fig1_electricity_access.png")
plt.show()
```

```python
plt.figure(figsize = (10, 6))

sns.lineplot(data = data_focus,
             x = "Year",
             y = "Renewable energy share in the total final energy consumption (%)",
             hue = "Entity",
             marker = "o")

plt.title("Renewable Energy Share in Final Energy Consumption (20002020)")

plt.xticks(data_focus["Year"].unique(), rotation = 45)

plt.ylabel("Renewable Share (%)")
plt.xlabel("")

plt.grid()
plt.tight_layout()
plt.savefig("fig2_renewable_share.png")
plt.show()
```

```python
plt.figure(figsize = (10, 6))

plt.plot(data_focus['Year'], data_focus['Electricity from fossil fuels (TWh)'], label = 'Fossil Fuels')
plt.plot(data_focus['Year'], data_focus['Electricity from nuclear (TWh)'], label = 'Nuclear')
plt.plot(data_focus['Year'], data_focus['Electricity from renewables (TWh)'], label = 'Renewables')

plt.title("Electricity Generation by Source in India (20002020)")

plt.xticks(data_focus["Year"].unique(), rotation = 45)

plt.ylabel("Electricity (TWh)")
plt.xlabel("")

plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig("fig3_energy_sources.png")
plt.show()
```
